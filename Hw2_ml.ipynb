The neuron doctrine emerges from Cajal's ideas at the end of the 19th century but not only.
A good amount of researchers seemed to discover similar results in different areas while using different
methodologies. 
The neuron doctrine revealed that the nervous system is made of discrete invidual cells, functionning 
independently while communicating with the axons. 
In his paper, Shepherd's assesses a few points that may need to be revisited. We could describe his view
as a more dynamic and type-specific interactions. Shepherd states that the nervous system is a cluster
of areas that interact together based on the function they need to accomplish. He also stipulates that neurons
may have different synaptic positions based on the task they need to do. Finally, Shepherd concludes by saying
that these functional units are a complex system of interactions that overlap at every level of the nervous system.
By comparing these two approaches, we realize that Cajal's approach can be compared to a more classical approach of
programming where neurons would react only when there is a specific input. The all-or-none approach represents how classical programming is at its best.
On the other hand, Shepherd's view focuses more on the idea that connections can be made and expanded in a dynamic approach. The brain or neural network learns from the input and creates connections that will be remembered in the future. Adding to this view, we could be talking of interconnections because these connections grow with each other but can also be used in different angles.
Finally, the neural network approach or Shepherd's view reflect more complex behaviors that grow dynamically while the more classical view focuses on the all-or-none state of neurons or inputs. 



```python
#Question 6

from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_classification


#we are using the sklearn perceptron algorithm, which is a subclass of SGDClassifier
X, y = load_digits(return_X_y=True) #loading data
print("Data:")
print(X)
print(y)
print()
clf = Perceptron(tol=1e-3, random_state=0)


Perceptron()
clf.fit(X, y) #fitting the data using a gradient descent
print("Using a gradient descent:")
print(clf.score(X,y))
print()

print("Using a gradient descent with only 1 iteration:")
clf.partial_fit(X,y)
print(clf.score(X, y))
```
